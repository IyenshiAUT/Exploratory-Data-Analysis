## Data Cleaning

```python
df_copy = df.copy()
```

```python
df_copy.drop(df_copy.index[10472], inplace=True)
```

```python
# convert into int
df_copy['Reviews'] = df_copy['Reviews'].astype(int)
```

```python
df_copy.info()
```

```python
df['Size'].unique()
```

```python
df_copy['Size'] = df_copy['Size'].str.replace('M', '000')
df_copy['Size'] = df_copy['Size'].str.replace('k', '')
df_copy['Size'] = df_copy['Size'].replace('Varies with device', np.nan)
df_copy['Size'] = df_copy['Size'].astype(float)
```

```python
df_copy['Size']
```

```python
df.head()
```

```python
df['Installs'].unique()
```

```python
df['Price'].unique()
```

```python
char_removed = ['+',',','$']
cols_to_clean = ['Installs', 'Price']
for item in char_removed:
    for col in cols_to_clean:
        df_copy[col] = df_copy[col].str.replace(item, '')
```

```python
df_copy['Installs'].unique()
```

```python
df_copy['Installs'] = df_copy['Installs'].astype(int)
df_copy['Price'] = df_copy['Price'].astype(float)
```

```python
df_copy.info()
```

```python
df_copy['Last Updated'].unique()
```

```python
df_copy['Last Updated'] = pd.to_datetime(df_copy['Last Updated'])
df_copy['Day'] = df_copy['Last Updated'].dt.day
df_copy['Month'] = df_copy['Last Updated'].dt.month
df_copy['Year'] = df_copy['Last Updated'].dt.year
df_copy.drop('Last Updated', axis =1, inplace=True)
```

```python
df_copy.head()
```

```python
df_copy.to_csv("Google Data _Halfly_ Cleaned.csv", index=False)
```

## Explore Data

```python
numeric_features = [feature for feature in df_copy.columns if df_copy[feature].dtype != 'O']
cat_features = [feature for feature in df_copy.columns if df_copy[feature].dtype == 'O']

print(f'No of numeric features : {len(numeric_features)},{numeric_features}')
print(f'No of categorical features : {len(cat_features)}, {cat_features}')
```

```python
for column in cat_features:
    print(df[column].value_counts(normalize=True))
    print('__________________________')
```

```python
# proportion of count as plot
plt.figure(figsize=(15, 15))
plt.suptitle("Univariate Analysis of Numerical Features", fontsize=20, fontweight='bold', alpha=0.8)

for i in range(0, len(numeric_features)):
    plt.subplot(5, 3, i+1)
    sns.kdeplot(x=df_copy[numeric_features[i]], shade= True, color='r')
    plt.xlabel(numeric_features[i])
    plt.tight_layout()
```

Observations

- Rating and Year are left skewed while Reviewa, Size, Installs are Right Skewed.
